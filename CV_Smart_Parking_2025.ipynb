### Plate Number Detection & Extraction

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'

import easyocr
import matplotlib.pyplot as plt
```

```python
# Load the image
image = cv2.imread("car4.jpg")

# Convert to gray scale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Load the pre-trained Haar Cascade model for plate detection
model = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_russian_plate_number.xml")

# Detect plates in the image
plates = model.detectMultiScale(gray, scaleFactor=1.1, minSize=(30,30), minNeighbors=5)

# Loop through the detected plates and draw bounding boxes
for (x,y,w,h) in plates:
    cv2.rectangle(image, (x,y), (x+w, y+h), (0, 0, 255), 2)

# Display the image with the detected plates highlighted
plt.imshow(image)
plt.axis("off")
plt.show()

# Crop the detected plate region
cropped_image = image[y:y+h, x:x+w]
plt.imshow(cropped_image)
plt.axis("off")
plt.show()
```

```python
# Perform OCR on the cropped region
reader = easyocr.Reader(["en"])
cropped_image = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)
results = reader.readtext(cropped_image, detail=0)[-1]
print(f"The plate number on this vehicle is {results}")
```

### Pre-processing

```python
image = cv2.imread("car4.jpg")

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Denoise the image using Gaussian Blur
gray = cv2.GaussianBlur(gray, (1, 1), 0)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(1, 2)
ax[0].imshow(image_rgb)
ax[0].axis("off")
ax[0].set_title("Original Image")
ax[1].imshow(gray, cmap='gray')
ax[1].set_title("Pre-processed Image")
ax[1].axis("off")
plt.show()

model = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_russian_plate_number.xml")
plates = model.detectMultiScale(gray, scaleFactor=1.1, minSize=(30,30), minNeighbors=5)
for (x,y,w,h) in plates:
    cv2.rectangle(image, (x,y), (x+w, y+h), (0, 0, 255), 2)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.title("Plate Detection")
plt.axis("off")
plt.show()


# Crop the image using the bounding box
cropped_image = image_rgb[y:y+h, x:x+w]
plt.imshow(cropped_image, cmap='gray')
plt.title("Before processing")
plt.axis("off")
plt.show()

# Pre-process cropped plate number
cropped_image = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)
# Apply thresholding (Otsu's method)
_, cropped_image = cv2.threshold(cropped_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.imshow(cropped_image, cmap='gray')
plt.title("After processing")
plt.axis("off")
plt.show()

# Use EasyOCR to extract text from the cropped image
## The First time of uing OCR, you'll be required to download somefiles
## This won't take long, depending on the quality of your network
## Subsequently, you won't need to download them anymore

reader = easyocr.Reader(["en"])
results = reader.readtext(cropped_image, detail=0)
print(f"The plate number on this vehicle is {results}")
```
